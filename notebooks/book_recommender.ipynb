## Popularity Based Recommendation 

```python
ratings_df.sample(3)
```

```python
books_df.sample(3)
```

```python
books_ratings_df = pd.merge(books_df, ratings_df, on='ISBN', how='inner')
print(books_ratings_df.shape)
```

```python
books_ratings_df.sample(3)
```

```python
number_of_ratings = books_ratings_df['Book-Title'].value_counts().reset_index()
number_of_ratings.columns = ['Book-Title', 'num_ratings']
number_of_ratings.head()
```

```python
average_ratings = books_ratings_df.groupby('Book-Title')['Book-Rating'].mean().reset_index()
average_ratings.columns = ['Book-Title', 'avg_ratings']
average_ratings.head()
```

```python
top_books_df = pd.merge(number_of_ratings, average_ratings, on='Book-Title')
print(number_of_ratings.shape)
print(average_ratings.shape)
print(top_books_df.shape)
top_books_df.head()
```

```python
MIN_RATINGS_REQUIRED = 250
TOP_N = 50
```

```python
top_N_books_df = top_books_df[top_books_df['num_ratings'] >= MIN_RATINGS_REQUIRED].sort_values(
    ['avg_ratings'], ascending=False).head(TOP_N)
```

```python
top_N_books_df.head()
```

```python
# merge to get all the details for top N books
top_N_books_df = top_N_books_df.merge(
    books_df.drop(columns=['ISBN']).drop_duplicates(subset=['Book-Title']),
    on='Book-Title',
    how='left'
)
print(top_N_books_df.shape)
top_N_books_df.head()
```

## Collaborative Filtering Based Recommendation 

```python
MIN_RATING_PER_USER = 250
MIN_BOOK_RATING = 50
```

```python
ratings_per_user = books_ratings_df['User-ID'].value_counts()
top_raters = ratings_per_user >= MIN_RATING_PER_USER
top_raters = top_raters[top_raters].index
```

```python
book_rating_frequency = books_ratings_df[books_ratings_df['User-ID'].isin(top_raters)][
    'Book-Title'].value_counts()
top_books = book_rating_frequency >= MIN_BOOK_RATING
top_books = top_books[top_books].index
```

```python
filtered_books = books_ratings_df[
    (books_ratings_df['User-ID'].isin(top_raters)) &
    (books_ratings_df['Book-Title'].isin(top_books))
]
```

```python
book_rating_matrix = filtered_books.pivot_table(
    index='Book-Title', columns='User-ID', values='Book-Rating')
book_rating_matrix = book_rating_matrix.fillna(0.)
```

```python
book_rating_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(book_rating_matrix)
```

```python
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend_book(book_name):
    index = np.where(book_rating_matrix.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books_df[books_df['Book-Title'] == book_rating_matrix.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        img_url = temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values[0]
        img_url = 'https' + img_url[4:]
        item.append(img_url)

        data.append(item)

    return data
```

```python
recommend_book('1984')
```

```python
recommend_book('A Painted House')
```
